### Bull Call Spread

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
def call_payoff(sT, strike_price, premium):
    return np.where(sT > strike_price, sT - strike_price, 0) - premium
```

```python
# Reliance stock price 
spot_price = 1457

# Long call
strike_price_long_call = 1460 
premium_long_call = 42.5

# Short call
strike_price_short_call = 1560
premium_short_call = 29.5

# Stock price range at expiration of the call
sT = np.arange(0.95*spot_price,1.1*spot_price,1)
```

Long 1460 strike call payoff

```python
payoff_long_call = call_payoff(sT, strike_price_long_call, premium_long_call)
# Plot
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(sT,payoff_long_call,label='Long 1460 Strike Call',color='g')
plt.xlabel('Reliance Stock Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

Short 1560 call payoff

```python
payoff_short_call = call_payoff(sT, strike_price_short_call, premium_short_call) * -1.0
# Plot
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(sT,payoff_short_call,label='Short 1560 Strike Call',color='r')
plt.xlabel('Reliance Stock Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

Bull Call Spread Payoff

```python
payoff_bull_call_spread = payoff_long_call + payoff_short_call

print ("Max Profit:", max(payoff_bull_call_spread))
print ("Max Loss:", min(payoff_bull_call_spread))
# Plot
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(sT,payoff_long_call,'--',label='Long 1460 Strike Call',color='g')
ax.plot(sT,payoff_short_call,'--',label='Short 1560 Strike Call ',color='r')
ax.plot(sT,payoff_bull_call_spread,label='Bull Call Spread')
plt.xlabel('Reliance Stock Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

###Covered Call

```python
def call_payoff(sT, strike_price, premium):
    return np.where(sT > strike_price, sT - strike_price, 0) - premium
```

```python
# Reliance stock price 
spot_price = 1457 

# Short call
strike_price_short_call = 1457
premium_short_call = 35

# Stock price range at expiration of the call
sT = np.arange(0.9*spot_price,1.1*spot_price,1)
```

```python
payoff_stock = sT-spot_price
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(sT,payoff_stock,label='Reliance Stock Payoff',color='g')
plt.xlabel('Reliance Stock Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

```python
payoff_short_call = call_payoff(sT, strike_price_short_call, premium_short_call) * -1.0
# Plot
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(sT,payoff_short_call,label='Short 1457 Strike Call',color='r')
plt.xlabel('Reliance Stock Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

```python
covered_call_payoff = payoff_stock + payoff_short_call
# Plot
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(sT,payoff_stock,'--',label='Long Reliance Stock',color='g')
ax.plot(sT,payoff_short_call,'--',label='Short 1457 Strike Call ',color='r')
ax.plot(sT,covered_call_payoff,label='Covered Call')
plt.xlabel('Reliance Stock Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()

```

###Protective Put

```python
def put_payoff(sT, strike_price, premium):
    return np.where(sT < strike_price, strike_price - sT, 0) - premium
```

```python
# Reliance stock price 
spot_price = 1457

# stock purchase price
stock_purchase_price = 1457

# Short put
strike_price_long_put = 1457
premium_long_put = 35

# Stock price range at expiration of the put
sT = np.arange(0.9*spot_price,1.1*spot_price,1)
```

```python
payoff_long_put = put_payoff(sT, strike_price_long_put, premium_long_put)
# Plot
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(sT,payoff_long_put,label='Long 1457 Strike Put',color='r')
plt.xlabel('Reliance Stock Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

```python
payoff_auro_pharma_stock = sT - stock_purchase_price
# Plot
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(sT,payoff_auro_pharma_stock,label='Long Reliance Stock',color='g')
plt.xlabel('Reliance Stock Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

```python
payoff_protective_put = payoff_auro_pharma_stock + payoff_long_put

# Plot
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(sT,payoff_auro_pharma_stock,'--',label='Long Reliance Stock',color='g')
ax.plot(sT,payoff_long_put,'--',label='Long 1457 Strike Put',color='r')
ax.plot(sT,payoff_protective_put,label='Protective Put')
plt.xlabel('Reliance Stock Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```
